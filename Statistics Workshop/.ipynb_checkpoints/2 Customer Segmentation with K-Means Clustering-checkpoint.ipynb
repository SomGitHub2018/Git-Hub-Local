```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Load customer data
df = pd.read_csv('customer_data.csv')
X = df[['Annual_Income', 'Spending_Score']]
X.head()
```

```python
# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
df['Segment'] = kmeans.fit_predict(X)

# Visualize clusters
plt.scatter(X['Annual_Income'], X['Spending_Score'], c=df['Segment'], cmap='viridis')
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation')
plt.show()
```

### Choosing clusters
wcss -> within clusters sum of squares

finding wcss values for different number of clusters

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters= i, init = 'k-means++', random_state=42)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('Elbow graph')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Step 4: Analyze cluster characteristics
segment_summary = df.groupby('Segment')[['Annual_Income', 'Spending_Score']].mean().round(2)
print("Segment Summary:\n", segment_summary)
```

```python
# Step 5: Identify high-value segment
high_value_segment = segment_summary[(segment_summary['Annual_Income'] > segment_summary['Annual_Income'].mean()) &
                                     (segment_summary['Spending_Score'] > segment_summary['Spending_Score'].mean())]

print("\nHigh-Value Segment(s):")
print(high_value_segment)
```

Explaination
- Standardizes the data to ensure fair clustering. 

- Clusters customers into 4 groups based on income and spending score.

- Visualizes the clusters for intuitive understanding.

- Summarizes each segments average income and spending.

- Identifies high-value segmentsthose with above-average income and spending
